```python
df.nsmallest(5, 'floors')
```

```python
df.nlargest(5, 'floors')
```

By comparing the appartments with 3.5 floors with apartments with the minimum as 1 floor only, we can see that it doesn't have a real impact on the pricing, however it is clear that the sqft_living is the attribute that is affecting the pricing much.
As comparing app nb 4 to that nb 11594 of similar sqft_living, nb of bedrooms, nb of bathrooms, 0 waterfronts both, same view/grade.. but the first has only one floor and the second has 3.5 floor, we can see that there isnt much difference in the pricing, meaning that the increment of floors didn't impact the pricing. 

```python
df.nlargest(10, 'sqft_living')
```

```python
df.nsmallest(10, 'sqft_living')
```

It is clear for sure that the sqft_living is highly affecting the pricing of the appartments, and that is approximately by a factor of 10. It also impacts the nb of bedrooms and bathrooms inside each apartment, as the square feet of living space increases the nb of bathrooms and bedrooms increase proportionally. This indicates that higher living areas result in higher prices.


```python
df.nlargest(5, 'price')
```

```python
df.nsmallest(5, 'price')
```

While comparing the highest and lowest prices, we may say that square feet of living space has the highest effect on the price on the appartment. With the increase of the square feet of living space, the number of bathrooms and bedrooms increase correspondingly. The grade the appartment is on also has an impact, higher grades have higher prices. Also, the sqaure feet of basement has an impact as well, where the appartments with 0 square feet of basement space are of the lowest prices.
